# ¿Qué es un "Modern Data Stack"?
Un "modern data stack" (conjunto de datos moderno) es un enfoque contemporáneo para gestionar y analizar datos utilizando una combinación de herramientas y tecnologías modernas, generalmente basadas en la nube. ## 1. Fuentes de Datos (Sources)

Las fuentes de datos son los orígenes desde los cuales se extraen los datos. Pueden ser internas o externas y presentan datos en diversos formatos y estructuras.

Tipos de Fuentes:

- Bases de Datos Relacionales: Como PostgreSQL, MySQL.
- APIs: Servicios web que proporcionan datos en tiempo real.
- Archivos: CSV, JSON, XML, hojas de cálculo.
- Sistemas de Streaming: Kafka, Kinesis.
- Dispositivos IoT y Sensores: Generan datos en tiempo real.
- Aplicaciones Empresariales: CRM, ERP, sistemas de gestión.

## 1'. Bases de datos transaccionales (OLTP)
Una base de datos transaccional es un tipo de sistema de gestión de bases de datos diseñado para manejar y optimizar las operaciones transaccionales, que implican operaciones frecuentes de lectura y escritura, como insertar, actualizar, eliminar y recuperar datos en tiempo real. Estas bases de datos están enfocadas en garantizar la consistencia y eficiencia de las transacciones de negocios y son esenciales para aplicaciones que requieren alta concurrencia y acceso rápido a datos.

Características clave de una base de datos transaccional:
### 1. Soporte para transacciones ACID:

- Atomicidad: Una transacción es atómica, lo que significa que todas las operaciones dentro de ella deben completarse exitosamente o ninguna se ejecuta. Esto garantiza la integridad de los datos.

- Consistencia: Las transacciones llevan a la base de datos de un estado válido a otro, manteniendo las reglas de integridad de los datos.

- Aislamiento: Las transacciones simultáneas se ejecutan como si fueran secuenciales, evitando interferencias entre ellas.

- Durabilidad: Una vez que una transacción se confirma, sus efectos permanecen incluso en caso de fallos en el sistema.

2. Alta concurrencia: Estas bases de datos están diseñadas para manejar múltiples operaciones simultáneamente sin que los usuarios interfieran entre sí, lo que es crucial en aplicaciones con gran número de usuarios concurrentes (por ejemplo, sistemas bancarios o de comercio electrónico).

3. Optimización para operaciones de escritura: Las bases de datos transaccionales están optimizadas para manejar grandes volúmenes de operaciones de escritura en tiempo real, como insertar, actualizar y eliminar datos. Esto es típico en aplicaciones que necesitan registrar continuamente eventos o transacciones.

4. Integridad referencial: Implementan mecanismos como claves foráneas y restricciones para asegurar que las relaciones entre las tablas y los datos se mantengan coherentes, evitando datos huérfanos o duplicados.

5. Modelado basado en filas: Las bases de datos transaccionales generalmente almacenan los datos en formato de filas, lo que permite acceder rápidamente a todos los datos relacionados con una transacción específica. Este enfoque es ideal para operaciones transaccionales, pero no para análisis masivos.


### ¿Por qué vamos a ver PostgreSQL como herramienta de OLTP?

- Open source y enorme soporte

- Compatibilidad con datos estructurados y no estructurados: Soporta tanto datos relacionales como JSON/JSONB, permitiendo manejar datos semi-estructurados de manera eficiente, lo que lo hace versátil frente a bases de datos que solo gestionan datos estructurados.

- Soporte avanzado para transacciones ACID: Implementa sólidamente las propiedades ACID (Atomicidad, Consistencia, Aislamiento y Durabilidad), lo que garantiza la integridad de las transacciones y facilita la alta concurrencia sin bloqueos excesivos.

- Extensibilidad y personalización: Ofrece soporte para extensiones y personalización a través de funciones definidas por el usuario y procedimientos almacenados, lo que permite adaptarlo a diversas necesidades sin necesidad de complementos comerciales.

- Seguridad avanzada: Proporciona características avanzadas de seguridad, como autenticación a nivel de fila y permisos granulares, lo que garantiza un control detallado del acceso a los datos, algo que otras bases de datos no siempre ofrecen de forma nativa.

## 2. Ingesta de Datos (Ingestion)

La ingestia de datos es el proceso de extraer datos de las fuentes y cargarlos en un sistema centralizado, una base de datos o un data lake.

### ¿Por qué es necesario aprender alguna solución de ingesta?
- Manejo de datos del mundo real: Los datos raramente están en un formato listo para el análisis. Las herramientas de ingesta permiten extraer, transformar y cargar datos desde múltiples fuentes (APIs, bases de datos, archivos CSV, etc.) en un formato usable.

- Automatización de procesos: Los data scientists a menudo necesitan automatizar la recolección de datos de forma recurrente. Las herramientas de ingesta facilitan este proceso mediante flujos de trabajo automatizados y repetibles.

- Limpieza y transformación de datos: Durante la ingesta, los datos pueden limpiarse y transformarse de acuerdo con las necesidades del análisis. Esto asegura que los modelos trabajen con datos de calidad y bien estructurados.

- Integración con pipelines: Las herramientas de ingesta son una parte importante de los pipelines de datos, que son flujos de trabajo automatizados que gestionan el movimiento de datos desde la fuente hasta el análisis o el modelado.

- Escalabilidad: A medida que los proyectos de ciencia de datos crecen, las herramientas de ingesta permiten manejar grandes volúmenes de datos de forma eficiente, algo crucial para trabajar en proyectos de big data.

- Facilitación del análisis: Tener una buena estructura de ingesta reduce el tiempo necesario para preparar los datos, dejando más tiempo para el análisis, modelado y extracción de insights.

##  3. Data Warehouse (OLAP)

Un data warehouse (almacén de datos) es un sistema especializado para el almacenamiento, integración y análisis de grandes volúmenes de datos históricos provenientes de diversas fuentes. Está diseñado específicamente para realizar consultas analíticas complejas y generar informes que apoyan la toma de decisiones empresariales. A diferencia de las bases de datos transaccionales (OLTP), un data warehouse se enfoca en el análisis de datos (OLAP, Online Analytical Processing), optimizando el rendimiento de consultas de lectura intensiva

### ¿Por qué vamos a ver DuckDB/Motherduck como herramienta OLAP?

- Almacenamiento columnar: Optimiza el rendimiento de las consultas analíticas al almacenar los datos por columnas, lo que reduce la lectura de datos innecesarios y mejora la velocidad en análisis de grandes volúmenes.

- Ejecución embebida: Se puede integrar directamente en aplicaciones y entornos de desarrollo como Python o R sin necesidad de infraestructura o servidores externos, lo que facilita su uso en flujos de trabajo de ciencia de datos.

- Rendimiento eficiente: Está diseñado para manejar consultas complejas y grandes volúmenes de datos de manera rápida, aprovechando formatos de archivo como Parquet y Arrow para procesar datos de manera eficiente.

- Compatibilidad con entornos de ciencia de datos: Se integra fácilmente con lenguajes y herramientas populares en ciencia de datos, como Pandas, Python, R y Jupyter Notebooks, permitiendo un análisis fluido dentro de los ecosistemas ya utilizados por los data scientists.

- Fácil de usar y sin infraestructura compleja: No requiere configuración de servidores o infraestructura avanzada, lo que lo convierte en una opción ideal para análisis exploratorio y rápido sin complicaciones técnicas.
